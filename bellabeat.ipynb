### 3.7 Data Transformation
To begin, I will convert the "ActivityDate" in the daily_activity dataset, "ActivityHour" in both the hourly_steps and hourly_calories datasets, and "SleepDay" in the sleepday dataset into the datetime data type. This will ensure accurate and efficient processing of time-series data.

```python
daily_activity['ActivityDate'] = pd.to_datetime(daily_activity['ActivityDate'])
daily_intensity['ActivityDay'] = pd.to_datetime(daily_intensity['ActivityDay'])
daily_steps['ActivityDay'] = pd.to_datetime(daily_steps['ActivityDay'])
daily_calories['ActivityDay'] = pd.to_datetime(daily_calories['ActivityDay'])
hourly_steps['ActivityHour'] = pd.to_datetime(hourly_steps['ActivityHour'])
hourly_calories['ActivityHour'] = pd.to_datetime(hourly_calories['ActivityHour'])
sleep_day['SleepDay'] = pd.to_datetime(sleep_day['SleepDay'])
weight_log['Date'] = pd.to_datetime(weight_log['Date'])
```

```python
print(daily_activity['ActivityDate'].dtypes)
print(daily_intensity['ActivityDay'].dtypes)
print(daily_steps['ActivityDay'].dtypes)
print(daily_calories['ActivityDay'].dtypes)
print(hourly_steps['ActivityHour'].dtypes)
print(hourly_calories['ActivityHour'].dtypes)
print(sleep_day['SleepDay'].dtypes)
print(weight_log['Date'].dtypes)
```

Let's see the average steps of everyday of the week

```python
days_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
daily_activity['Day'] = pd.Categorical(daily_activity['Day'], categories=days_order, ordered=True)
average_steps = daily_activity.groupby('Day')['TotalSteps'].mean().reset_index()

sns.set(style='whitegrid')

plt.figure(figsize=(10,8))
bar_plot = sns.barplot(x='Day', y='TotalSteps', data=average_steps, palette='coolwarm')
plt.title('Average steps by day of the week')
plt.xlabel('Day of the week')
plt.ylabel('Average Steps')
plt.xticks(rotation=45)
for p in bar_plot.patches:
    bar_plot.annotate(format(p.get_height(), '.1f'),  
                      (p.get_x() + p.get_width() / 2., p.get_height()), 
                      ha = 'center', 
                      va = 'center', 
                      xytext = (0, 10),
                      textcoords = 'offset points',
                      rotation=45)
plt.show()
```

#### Average Steps by Day of the Week
This chart shows the average number of steps people take each day of the week. Saturday has the highest steps at 8,153, meaning people are most active on weekends. The lowest steps are on Sunday, at 6,933, suggesting that people usually relax more on this day.

```python
average_calories = daily_activity.groupby('Day')['Calories'].mean().reset_index()

sns.set(style='whitegrid')
plt.figure(figsize=(10,8))
bar_plot_c = sns.barplot(x='Day', y='Calories',data=average_calories, palette='coolwarm')
plt.title('Average calories burned by day of the week')
plt.xlabel('Day of the week')
plt.ylabel('Average calories burned')
plt.xticks(rotation=45)
for p in bar_plot_c.patches:
    bar_plot_c.annotate(format(p.get_height(), '.1f'),  
                      (p.get_x() + p.get_width() / 2., p.get_height()), 
                      ha = 'center', 
                      va = 'center', 
                      xytext = (0, 10),
                      textcoords = 'offset points',
                      rotation=45)
plt.show()
```

#### Average Calories Burned by Day of the Week
This chart shows the average calories burned by people on different days of the week. It matches well with the steps chart, where weekdays generally have a higher calorie burn compared to weekends, with Tuesday being the highest. This means that more active days typically lead to burning more calories.

```python
hourly_avg = hourly_activity.groupby('Hour').agg({'StepTotal':'mean','Calories':'mean'}).reset_index()


sns.set(style='whitegrid')
plt.figure(figsize=(10,8))
bar_width = 0.35
index = hourly_avg['Hour']
bar1 = plt.bar(index - bar_width/2, hourly_avg['StepTotal'], bar_width, label='Average Steps', color='b')
bar2 = plt.bar(index + bar_width/2, hourly_avg['Calories'], bar_width, label='Average Calories', color='g')
plt.xlabel('Hour of the Day')
plt.ylabel('Average Values')
plt.title('Average Steps and Calories Burned per Hour')
plt.xticks(index)
plt.legend()
plt.show()
```

#### Average Steps and Calories Burned per Hour
This bar chart displays the average steps and calories burned by people during each hour of the day. You can see that people are most active between 5 PM and 7 PM, likely because they are moving more after work or school. Calories burned also increase during these hours, but the increase isn't as sharp as the increase in steps.

```python
daily_avg = daily_activity.groupby('Day').agg({'TotalActivityMinutes':'mean',
    'VeryActiveMinutes': 'mean',
    'FairlyActiveMinutes': 'mean',
    'LightlyActiveMinutes': 'mean',
    'SedentaryMinutes': 'mean'}).reset_index()

sns.set(style='whitegrid')
plt.figure(figsize=(10,8))
plt.plot(daily_avg['Day'], daily_avg['VeryActiveMinutes'], label='Very Activity Minutes', marker='o')
plt.plot(daily_avg['Day'], daily_avg['FairlyActiveMinutes'], label='Fairly Active Minutes', marker='o')
plt.plot(daily_avg['Day'], daily_avg['LightlyActiveMinutes'], label='Lightly Active Minutes', marker='o')
plt.plot(daily_avg['Day'], daily_avg['SedentaryMinutes'], label='Sedentary Minutes', marker='o')


plt.title('Average Activity Minutes per Day')
plt.xlabel('Day of the week')
plt.ylabel('Average Minutes')
plt.legend(title='Activity Type')
plt.xticks(rotation=45)
plt.show()
```

#### Average Activity Minutes per Day

This line chart shows how many minutes people are active or inactive each day of the week. People are mostly inactive (sedentary) every day, with inactive time staying high throughout the week. There's a small increase in very active minutes around the middle of the week, which lines up with the days when people take more steps and burn more calories. This suggests that the most of the people have some kind og desk job which requires less movement during the day.

```python
fig, ax = plt.subplots(figsize=(8,5))
x = np.array(["Sedentary Minutes", "TotalActiveMinutes"])
y = np.array([daily_activity["SedentaryMinutes"].mean(),
              daily_activity["TotalActivityMinutes"].mean()])

plt.title("Sedentary Minutes Vs Total Active Minutes")
plt.bar(x, y, width = 0.8, color = "b")
plt.show()
```

#### Sedentary Minutes Vs Total Active Minutes
This bar chart compares the time people spend inactive (sedentary) with the time they are active during the day. It clearly shows that people are inactive much more than they are active, which could be risky for their health. This suggests a need for ways to help people be more active every day.

```python
daily_dist = daily_activity.groupby('Day').agg({'TotalDistance': 'mean',
    'VeryActiveDistance': 'mean',
    'ModeratelyActiveDistance': 'mean',
    'LightActiveDistance': 'mean',
    'SedentaryActiveDistance': 'mean'}).reset_index()

sns.set(style='whitegrid')
plt.figure(figsize=(10,8))
plt.plot(daily_dist['Day'], daily_dist['VeryActiveDistance'], label='Very Active Distance', marker='o')
plt.plot(daily_dist['Day'], daily_dist['ModeratelyActiveDistance'], label='Moderately Active Distance', marker='o')
plt.plot(daily_dist['Day'], daily_dist['LightActiveDistance'], label='Light Active Distance', marker='o')
plt.plot(daily_dist['Day'], daily_dist['SedentaryActiveDistance'], label='Sedentary Active Distance', marker='o')


plt.title('Average Distance covered per Day')
plt.xlabel('Day of the week')
plt.ylabel('Average Distance')
plt.legend(title='Activity Type')
plt.xticks(rotation=45)
plt.show()
```

#### Average Distance Covered per Day
This line chart tracks the distances people cover each day based on how active they are: very active, moderately active, lightly active, and inactive. The distance covered when people are very active or moderately active stays about the same throughout the week. However, there's an increase in the distance covered when people are lightly active during the weekend.

```python
days_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
sleep_day['Day'] = pd.Categorical(sleep_day['Day'], categories=days_order, ordered=True)
avg_sleep = sleep_day.groupby('Day')['TotalMinutesAsleep'].mean().reset_index()
avg_sleep['TotalMinutesAsleep'] = avg_sleep['TotalMinutesAsleep'] / 60

sns.set(style='whitegrid')
plt.figure(figsize=(10,8))
bar_plot_s = sns.barplot(x='Day', y = 'TotalMinutesAsleep', data = avg_sleep, color='b')
plt.title('Average minutes slept in every day of the week')
plt.xlabel('Day of the week')
plt.ylabel('Minutes')
plt.xticks(rotation=45)
for p in bar_plot_s.patches:
    bar_plot_s.annotate(format(p.get_height(), '.1f'),  # Format the number to one decimal place
                      (p.get_x() + p.get_width() / 2., p.get_height()),  # Position for the text
                      ha = 'center',  # Center alignment
                      va = 'center',  # Center alignment vertically
                      xytext = (0, 10),  # Distance from the bar
                      textcoords = 'offset points',
                      rotation = 45)
plt.show()
```
